```python
plt.figure(figsize=(14,8))
ax = df[df.species=='Iris-setosa'].plot.scatter(x='sepal_length', y='sepal_width', label='Setosa')
df[df.species=='Iris-versicolor'].plot.scatter(x='sepal_length', y='sepal_width', color='red', label='Versicolor', ax=ax)
df[df.species=='Iris-virginica'].plot.scatter(x='sepal_length', y='sepal_width', color='orange', label='Virginica', ax=ax)
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
ax.set_title("Relationship between Sepal Length and Width")
plt.show()
```

```python
# Load the Iris dataset from Seaborn
iris = sns.load_dataset('iris')

# Create a scatterplot using Seaborn
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', style='species')

# Add a title
plt.title('Scatterplot of Sepal Length vs Sepal Width by Species')

# Show the plot
plt.show()
```

```python
# Histogram in PNG files.
for column in df:
    plt.figure()  # Create a new figure for each histogram.
    plt.hist(df[column], edgecolor='black', linewidth=1)  # Create histogram for column with specified edgecolor and linewidth.
    plt.title(f'Histogram of {column}')  # Set title for the histogram.
    plt.xlabel('Values')  # Set xlabel for the histogram.
    plt.ylabel('Frequency')  # Set ylabel for the histogram.
    plt.savefig(f'{column}_histogram.png', bbox_inches='tight')  # Save the histogram as a PNG file.
```

### Visualization Data:

* Python has lots of visualization tools from Pandas, Matplotlib and Seaborn Libraries. We'll use couples of them to visualize data. 
***

```python
# Load the Iris dataset
url = 'https://raw.githubusercontent.com/CarlosRigueti/pands-project/main/iris.csv'
df = pd.read_csv(url)

# Define colors for each histogram
colors = {'sepal_length': 'red', 'sepal_width': 'green', 'petal_length': 'blue', 'petal_width': 'orange'}

# Create histograms for each numerical column
for column, color in colors.items():
    plt.figure(figsize=(8, 6))
    plt.hist(df[column], color=color, edgecolor='black', linewidth=1)
    plt.title(f'Histogram of {column}')
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
# Load the Iris dataset
url = 'https://raw.githubusercontent.com/CarlosRigueti/pands-project/main/iris.csv'
df = pd.read_csv(url)

# Define colors for each feature
colors = {'sepal_length': 'red', 'sepal_width': 'green', 'petal_length': 'blue', 'petal_width': 'orange'}

# Set the figure size
plt.figure(figsize=(12, 8))  # Adjust the figsize as needed

# Generate boxplots for each feature split by species
for feature, color in colors.items():
    sns.boxplot(x='species', y=feature, data=df, color=color)

# Add title and labels
plt.title('Boxplot of Iris Dataset Features by Species')
plt.xlabel('Species')
plt.ylabel('Value')

# Show the plot
plt.show()
```

```python
# Create a pair plot to visualize relationships between variables
# - 'data=df': Specifies the DataFrame containing the dataset to be visualized.
sns.pairplot(data = df, hue = 'species', height = 3)
```

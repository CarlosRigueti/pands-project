```python
# Histogram in PNG files.
for column in df:
    plt.figure()  # Create a new figure for each histogram.
    plt.hist(df[column], edgecolor='black', linewidth=1)  # Create histogram for column with specified edgecolor and linewidth.
    plt.title(f'Histogram of {column}')  # Set title for the histogram.
    plt.xlabel('Values')  # Set xlabel for the histogram.
    plt.ylabel('Frequency')  # Set ylabel for the histogram.
    plt.savefig(f'{column}_histogram.png', bbox_inches='tight')  # Save the histogram as a PNG file.
```

```python
# Load the Iris dataset from a URL into a DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')

# Define colors for each species
colors = {'setosa': 'lightyellow', 'versicolor': 'orange', 'virginica': 'lightblue'}

# Loop through each pair of variables
for i in range(len(df.columns) - 1):
    for j in range(i + 1, len(df.columns) - 1):
        # Create a new figure for each scatter plot
        plt.figure()
        
        # Plot each species with a different color
        for species, color in colors.items():
            # Filter data for the current species
            species_data = df[df['species'] == species]
            # Scatter plot for the current species with specified color
            plt.scatter(species_data.iloc[:, i], species_data.iloc[:, j], label=species, color=color)
        
        # Set title and labels for the scatter plot
        plt.title(f'Scatter Plot of {df.columns[i]} vs {df.columns[j]}')
        plt.xlabel(df.columns[i])
        plt.ylabel(df.columns[j])
        
        # Add legend to the scatter plot
        plt.legend()
        
        # Save the scatter plot as a PNG file
        plt.savefig(f'{df.columns[i]}_{df.columns[j]}_scatter.png', bbox_inches='tight')

print("Scatter plots with different colors for each species saved as PNG files.")
```

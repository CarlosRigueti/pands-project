```python
# Histogram in PNG files.
for column in df.columns:
    plt.figure()  # histogram for each figure.
    plt.hist(df[column], edgecolor='black', linewidth=1)
    plt.title(f'Histogram of {column}')
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.savefig(f'{column}_histogram.png', bbox_inches='tight')
```

```python
# Output a scatter plot for each variables.
num_cols = len(df.columns)
for i in range(num_cols):
    for j in range(i + 1, num_cols):
        plt.figure()  # Figure.
        plt.scatter(df.iloc[:, i], df.iloc[:, j]) # scatter.
        plt.title(f'Scatter Plot of {df.columns[i]} vs {df.columns[j]}') # title
        plt.xlabel(df.columns[i]) # xlabel
        plt.ylabel(df.columns[j]) # ylabel
        plt.savefig(f'{df.columns[i]}_{df.columns[j]}_scatter.png', bbox_inches='tight') # savefig

print("Histograms and scatter plots in PNG files.")
```

### Two Variable Plots:
***

```python
# Get just the petal lengths.
plen = df['petal_length']

# Show.
print(plen)

# Type.
print(type(plen))
```

```python
# Just get the numpy array.
plen = plen.to_numpy()

# Show.
plen
```

```python
# Petal widths.
pwidth = df['petal_width'].to_numpy()

# Show.
pwidth
```

```python
# Simple plot.
plt.plot(plen, pwidth, 'x')

# Axis labels.
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')

# Title.
plt.title('Iris Data Set')

# X limits.
plt.xlim(0, 8)

# Y limits.
plt.ylim(0, 4)
```

```python
# Create a new figure and set of axes.
fig, ax = plt.subplots()

# Simple plot.
ax.plot(plen, pwidth, 'x')

# Axis labels.
ax.set_xlabel('Petal Length (cm)')
ax.set_ylabel('Petal Width (cm)')

# Title.
ax.set_title('Iris Data Set')

# X limits.
ax.set_xlim(0, 8)

# Y limits.
ax.set_ylim(0, 4)
```

### Add a Best Fit Line:
***

$ y = mx + c = p_1 x^1 + p_0 = p_1 x + p_0$

```python
# Fit a straight line between x and y.
m, c = np.polyfit(plen, pwidth, 1)

# Show m and c.
m, c
```

```python
# Create a new figure and set of axes.
fig, ax = plt.subplots()

# Simple plot.
ax.plot(plen, pwidth, 'x')

# Simple plot.
ax.plot(plen, m * plen + c, 'r-')

# Axis labels.
ax.set_xlabel('Petal Length (cm)')
ax.set_ylabel('Petal Width (cm)')

# Title.
ax.set_title('Iris Data Set')

# X limits.
ax.set_xlim(0, 8)

# Y limits.
ax.set_ylim(0, 4)
```

```python
# x values for best fit line.
bf_x = np.linspace(0.0, 8.0, 100)

# y values for best fit line.
bf_y = m * bf_x + c
```

```python
# Create a new figure and set of axes.
fig, ax = plt.subplots()

# Simple plot.
ax.plot(plen, pwidth, 'x')

# Simple plot.
ax.plot(bf_x, bf_y, 'r-')

# Axis labels.
ax.set_xlabel('Petal Length (cm)')
ax.set_ylabel('Petal Width (cm)')

# Title.
ax.set_title('Iris Data Set')

# X limits.
ax.set_xlim(0, 8)

# Y limits.
ax.set_ylim(-1, 4)
```

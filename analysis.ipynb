### Data Visualization:


* Data visualization is the graphical representation of data to communicate information effectively and efficiently. It involves the use of visual elements such as charts, graphs, and maps to explore, analyze, and present data in a visual format.

***

```python
# Load the Iris dataset from a URL into a DataFrame
url = 'https://raw.githubusercontent.com/CarlosRigueti/pands-project/main/iris.csv'
df = pd.read_csv(url)

# Count the number of flowers for each species
species_counts = df['species'].value_counts()

# Plotting the bar chart
plt.bar(species_counts.index, species_counts.values, color=['lightblue', 'royalblue', 'navy'])

# Plotting the bar chart species
plt.xlabel('Species') 

# Plotting the bar chart Number of Flowers.
plt.ylabel('Number of Flowers')

# Plotting the bar chart Iris Flowers by Species.
plt.title('Number of Iris Flowers by Species') 

# Print the species counts bar chart in PNG.
plt.savefig('species_counts_bar_chart.png')

# Display the plt.
plt.show()
```

```python
# Create a new figure with specified size
plt.figure(figsize=(14,8))

# Plot scatter plot for Iris-setosa species, using sepal_length and sepal_width
ax = df[df.species == 'Iris-setosa'].plot.scatter(x='sepal_length', y='sepal_width', label='Setosa')

# Add scatter plot for Iris-versicolor species to the same axes
df[df.species == 'Iris-versicolor'].plot.scatter(x='sepal_length', y='sepal_width', color='lightblue', label='Versicolor', ax=ax)

# Add scatter plot for Iris-virginica species to the same axes
df[df.species == 'Iris-virginica'].plot.scatter(x='sepal_length', y='sepal_width', color='navy', label='Virginica', ax=ax)

# Set the label for the x-axis
ax.set_xlabel("Sepal Length")

# Set the label for the y-axis
ax.set_ylabel("Sepal Width")

# Set the title for the plot
ax.set_title("Relationship between Sepal Length and Width")

# Plotting the bar chart and saving as PNG
plt.savefig('sepal_length_vs_sepal_width_scatterplot.png')

# Display the plot
plt.show()
```

```python
# Load the Iris dataset from Seaborn
iris = sns.load_dataset('iris')

# Define colors for each species
colors = {'setosa': 'blue', 'versicolor': 'royalblue', 'virginica': 'navy'}

# Create a scatterplot using Seaborn
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', style='species', palette=colors)

# Add a title
plt.title('Scatterplot of Sepal Length vs Sepal Width by Species')

# Create a scatterplot using Seaborn and save as PNG
plt.savefig('seaborn_scatterplot.png')

# Show the plot
plt.show()
```

```python
# Define colors for each column
colors = {'sepal_length': 'lightblue', 'sepal_width': 'royalblue', 'petal_length': 'navy', 'petal_width': 'slateblue'}

# Histogram in PNG files
for column, color in colors.items():
    
    # Create a new figure for each histogram.
    plt.figure()  
    
    # Create histogram for column with specified edgecolor, linewidth, and color.
    plt.hist(df[column], color=color, edgecolor='black', linewidth=1) 
    
    # Set title for the histogram.
    plt.title(f'Histogram of {column}')  
    
    # Set xlabel for the histogram.
    plt.xlabel('Values')  
    
    # Set ylabel for the histogram.
    plt.ylabel('Frequency')  
    
    # Save the histogram as a PNG file.
    plt.savefig(f'{column}_histogram.png', bbox_inches='tight')  
    
    # Histograms and saving each as PNG
    plt.savefig(f'{column}_histogram.png', bbox_inches='tight')
    
    # Display the histogram
    plt.show()  
```

```python
# Load the Iris dataset from Seaborn
iris = sns.load_dataset('iris')

# Define a color palette
color_palette = ['blue', 'royalblue', 'navy']

# Create a boxplot using Seaborn with hue for differentiating by species and the specified palette
sns.boxplot(data=iris, x='species', y='sepal_length', hue='species', palette=color_palette)

# Add a title
plt.title('Boxplot of Sepal Length by Species')

# Boxplot using Seaborn and saving as PNG
plt.savefig('seaborn_boxplot.png')

# Display the plot
plt.show()

```

### Using a Heatmap:

***

"A heatmap is a valuable tool for visualizing the correlation between variables in a dataset. It offers several benefits:

* Quick Identification of Relationships: 

* Comprehensive Overview:

* Prioritization of Variables: 

* Effective Communication: 

```python
# Create a correlation matrix
correlation_matrix = df.groupby('species').corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Generate a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', fmt=".2f")

# Set the title
plt.title('Correlation Heatmap by Species')

# Correlation Heatmap and saving as PNG
plt.savefig('correlation_heatmap.png')

# Display the plot
plt.show()
```

```python
# Descibe the data set by Species. 
describe_species = df.groupby('species').describe().transpose()
print (describe_species)
```

```python
# Histogram in PNG files.
for column in df.columns:
    plt.figure()  # Create a new figure for each histogram.
    plt.hist(df[column], edgecolor='black', linewidth=1)  # Create histogram for column with specified edgecolor and linewidth.
    plt.title(f'Histogram of {column}')  # Set title for the histogram.
    plt.xlabel('Values')  # Set xlabel for the histogram.
    plt.ylabel('Frequency')  # Set ylabel for the histogram.
    plt.savefig(f'{column}_histogram.png', bbox_inches='tight')  # Save the histogram as a PNG file.
```

```python
# Output a scatter plot for each variables.
num_cols = len(df.columns)
for i in range(num_cols):
    for j in range(i + 1, num_cols):
        plt.figure()  # Figure.
        plt.scatter(df.iloc[:, i], df.iloc[:, j]) # scatter.
        plt.title(f'Scatter Plot of {df.columns[i]} vs {df.columns[j]}') # title
        plt.xlabel(df.columns[i]) # xlabel 
        plt.ylabel(df.columns[j]) # ylabel 
        plt.savefig(f'{df.columns[i]}_{df.columns[j]}_scatter.png', bbox_inches='tight') # savefig 

print("Histograms and scatter plots in PNG files.")
```

### Two Variable Plots:
***

```python
# Get just the petal lengths.[28]
plen = df['petal_length']

# Show. [29]
print(plen)

# Type. [30]
print(type(plen))
```

```python
# Just get the numpy array. [31]
plen = plen.to_numpy()

# Show.[32]
plen
```

```python
# Petal widths.[33]
pwidth = df['petal_width'].to_numpy()

# Show.[34]
pwidth
```

```python
# Simple plot.[35]
plt.plot(plen, pwidth, 'x')

# Axis labels.[36]
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')

# Title.[37]
plt.title('Iris Data Set')

# X limits.[38]
plt.xlim(0, 8)

# Y limits.[39]
plt.ylim(0, 4)
```

```python
# Create a new figure and set of axes.
fig, ax = plt.subplots()

# Simple plot.
ax.plot(plen, pwidth, 'x')

# Axis labels.
ax.set_xlabel('Petal Length (cm)')
ax.set_ylabel('Petal Width (cm)')

# Title.
ax.set_title('Iris Data Set')

# X limits.
ax.set_xlim(0, 8)

# Y limits.
ax.set_ylim(0, 4)
```
